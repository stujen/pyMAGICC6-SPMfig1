# Filtering and plotting

Here we provide a quick example of how to filter and plot with your data.

*Note: At the moment we have to import the Pyam library. In a later version, the data will come back with a data frame that can be filtered and plotted directly so we can halve the number of lines required.*

```python
# NBVAL_IGNORE_OUTPUT
import pymagicc
import pyam

from pymagicc import MAGICC6
from pymagicc import rcp26

import matplotlib.pyplot as plt  
plt.style.use('bmh') 
```

Run MAGICC.

```python
with MAGICC6() as magicc:
    results = magicc.run(rcp26)
```

Here we make our data Pyam compatible.

```python
results.df["model"] = "MAGICC6"
results.df["scenario"] = "RCP26"
results.df.rename({"time": "year"}, axis="columns", inplace=True)
```

```python
# NBVAL_IGNORE_OUTPUT
# Note that row ordering may vary
results.df.head()
```

```python
pyam_df = pyam.IamDataFrame(results.df)
```

Finally we filter and plot.

```python
pyam_df.filter(
    variable="*Concentrations|CO2"
).line_plot(figsize=(16, 9));
```

```python
pyam_df.filter(variable="*Temperature*").line_plot(figsize=(16, 9));
```
